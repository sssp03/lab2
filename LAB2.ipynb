```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate
from google.colab import drive
import os

# Montar Google Drive
drive.mount('/content/drive')

#ruta del archivo en Drive
file_path = "/content/drive/My Drive/lab 2/0284_001_004_EEG.hea"

# Verificar si el archivo existe
if os.path.exists(file_path):
    print("Archivo encontrado:", file_path)
else:
    print("Error: Archivo no encontrado.")

# Leer el contenido del archivo .hea
with open(file_path, 'r') as file:
    for line in file:
        print(line.strip())


#  CONVOLUCI√ìN


# c√≥digo y c√©dula
h1 = np.array([5, 6, 0, 0, 7, 1, 3])
x1 = np.array([1, 0, 5, 3, 4, 4, 2, 3, 9, 9])

h2 = np.array([5, 6, 0, 0, 7, 2, 1])
x2 = np.array([1, 0, 0, 0, 2, 4, 1, 8, 2, 2])

# Calcular la convoluci√≥n
y1 = np.convolve(x1, h1)
y2 = np.convolve(x2, h2)

# Graficar los resultados de la convoluci√≥n
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

axs[0].stem(y1)
axs[0].set_title("Convoluci√≥n de x1[n] y h1[n]")
axs[0].set_xlabel("n")
axs[0].set_ylabel("y1[n]")

axs[1].stem(y2)
axs[1].set_title("Convoluci√≥n de x2[n] y h2[n]")
axs[1].set_xlabel("n")
axs[1].set_ylabel("y2[n]")

plt.tight_layout()
plt.show()


#  CORRELACI√ìN ENTRE SE√ëALES


# Definir las se√±ales x1[n] y x2[n]
fs = 1 / (1.25e-3)  # Frecuencia de muestreo
n = np.arange(9)  # 9 muestras

x1_n = np.cos(2 * np.pi * 100 * n * (1.25e-3))
x2_n = np.sin(2 * np.pi * 100 * n * (1.25e-3))

# Calcular la correlaci√≥n cruzada
corr = correlate(x1_n, x2_n, mode='full')

# Graficar las se√±ales y la correlaci√≥n
fig, axs = plt.subplots(3, 1, figsize=(10, 8))

axs[0].stem(n, x1_n)
axs[0].set_title("Se√±al x1[n] = cos(2œÄ100nTs)")
axs[0].set_xlabel("n")
axs[0].set_ylabel("x1[n]")

axs[1].stem(n, x2_n)
axs[1].set_title("Se√±al x2[n] = sin(2œÄ100nTs)")
axs[1].set_xlabel("n")
axs[1].set_ylabel("x2[n]")

axs[2].stem(range(-len(n) + 1, len(n)), corr)
axs[2].set_title("Correlaci√≥n cruzada entre x1[n] y x2[n]")
axs[2].set_xlabel("Desplazamiento")
axs[2].set_ylabel("Correlaci√≥n")

plt.tight_layout()
plt.show()


# TRANSFORMADA DE FOURIER Y AN√ÅLISIS ESPECTRAL


fs = 1000  # Frecuencia de muestreo en Hz
t = np.linspace(0, 1, fs, endpoint=False)  # 1 segundo de datos
signal = np.sin(2 * np.pi * 50 * t) + 0.5 * np.sin(2 * np.pi * 120 * t)  # Se√±al con dos frecuencias

# Calcular la FFT
frequencies = np.fft.fftfreq(len(signal), 1/fs)
fft_values = np.fft.fft(signal)

# Graficar la se√±al original y su transformada de Fourier
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

axs[0].plot(t, signal)
axs[0].set_title("Se√±al en el tiempo")
axs[0].set_xlabel("Tiempo [s]")
axs[0].set_ylabel("[mV]")

axs[1].plot(frequencies[:fs//2], np.abs(fft_values[:fs//2]))  # Solo la parte positiva del espectro
axs[1].set_title("Espectro de frecuencia (FFT)")
axs[1].set_xlabel("Frecuencia [Hz]")
axs[1].set_ylabel("Magnitud")

plt.tight_layout()
plt.show()


# AN√ÅLISIS ESTAD√çSTICO DESCRIPTIVOS DE LA SE√ëAL


mean_freq = np.mean(frequencies[:fs//2])
median_freq = np.median(frequencies[:fs//2])
std_freq = np.std(frequencies[:fs//2])

# Graficar histograma de frecuencias
plt.figure(figsize=(8, 4))
plt.hist(frequencies[:fs//2], bins=30, alpha=0.75, color='b', edgecolor='black')
plt.title("Histograma de Frecuencias")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Muestras")
plt.show()

# Mostrar estad√≠sticas
print(f"Frecuencia media: {mean_freq:.2f} Hz")
print(f"Frecuencia mediana: {median_freq:.2f} Hz")
print(f"Desviaci√≥n est√°ndar: {std_freq:.2f} Hz")


# C√ÅLCULO DEL SNR


def calculate_snr(signal, noise):
    power_signal = np.mean(signal ** 2)
    power_noise = np.mean(noise ** 2)
    return power_signal / power_noise, 10 * np.log10(power_signal / power_noise)

def calculate_snr_variance(signal, noise):
    var_signal = np.var(signal)
    var_noise = np.var(noise)
    return var_signal / var_noise, 10 * np.log10(var_signal / var_noise)

def calculate_snr_energy(signal, noise):
    energy_signal = np.sum(signal ** 2)
    energy_noise = np.sum(noise ** 2)
    return energy_signal / energy_noise, 10 * np.log10(energy_signal / energy_noise)

def calculate_snr_rms(signal, noise):
    rms_signal = np.sqrt(np.mean(signal ** 2))
    rms_noise = np.sqrt(np.mean(noise ** 2))
    return (rms_signal / rms_noise) ** 2, 10 * np.log10((rms_signal / rms_noise) ** 2)

def calculate_snr_mean_std(signal, noise):
    mean_signal = np.mean(np.abs(signal))
    std_noise = np.std(noise)
    return mean_signal / std_noise, 10 * np.log10(mean_signal / std_noise)

def calculate_snr_peak(signal, noise):
    peak_signal = np.max(np.abs(signal))
    rms_noise = np.sqrt(np.mean(noise ** 2))
    return peak_signal / rms_noise, 10 * np.log10(peak_signal / rms_noise)

noise = 0.2 * np.random.randn(len(signal))
noisy_signal = signal + noise

snr, snr_db = calculate_snr(signal, noise)
snr_var, snr_var_db = calculate_snr_variance(signal, noise)
snr_energy, snr_energy_db = calculate_snr_energy(signal, noise)
snr_rms, snr_rms_db = calculate_snr_rms(signal, noise)
snr_mean_std, snr_mean_std_db = calculate_snr_mean_std(signal, noise)
snr_peak, snr_peak_db = calculate_snr_peak(signal, noise)

print(f"SNR (potencia): {snr:.2f}, SNR en dB: {snr_db:.2f} dB")
print(f"SNR (varianza): {snr_var:.2f}, SNR en dB: {snr_var_db:.2f} dB")
print(f"SNR (energ√≠a): {snr_energy:.2f}, SNR en dB: {snr_energy_db:.2f} dB")
print(f"SNR (RMS): {snr_rms:.2f}, SNR en dB: {snr_rms_db:.2f} dB")
print(f"SNR (Media/STD): {snr_mean_std:.2f}, SNR en dB: {snr_mean_std_db:.2f} dB")
print(f"SNR (Pico/RMS): {snr_peak:.2f}, SNR en dB: {snr_peak_db:.2f} dB")

plt.figure()
plt.plot(t, noisy_signal, label='Se√±al con ruido', alpha=0.6)
plt.plot(t, signal, label='Se√±al original', linestyle='dashed')
plt.legend()
plt.xlabel('Tiempo [s]')
plt.ylabel('[mV]')
plt.title('Se√±al Original vs. Se√±al con Ruido')
plt.show()
```
